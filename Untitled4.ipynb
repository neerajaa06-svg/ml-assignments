```python
def k_closest(points, k):
    points.sort(key=lambda p: p[0]**2 + p[1]**2)
    return points[:k]

# Example
points = [(1,3), (-2,2), (5,8), (0,1)]
k = 2

print(k_closest(points, k))
```
