```python
def group_anagrams(words):
    d = {}
    for w in words:
        k = ''.join(sorted(w))
        if k in d:
            d[k].append(w)
        else:
            d[k] = [w]
    return list(d.values())
print(group_anagrams(["listen", "silent", "enlist", "hello"]))
```

```python
def setZeroes(matrix):
    rows = len(matrix)
    cols = len(matrix[0]) 
    row_zero = [False] * rows
    col_zero = [False] * cols
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                row_zero[i] = True
                col_zero[j] = True
    for i in range(rows):
        for j in range(cols):
            if row_zero[i] or col_zero[j]:
                matrix[i][j] = 0
    return matrix
matrix = matrix = [[1,2,0],[4,5,6],[0,8,9]]
print(setZeroes(matrix))
print(setZeroes(matrix))
```

```python
def maxAverage(nums, k):
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for i in range(k, len(nums)):
        window_sum = window_sum + nums[i] - nums[i - k]
        if window_sum > max_sum:
            max_sum = window_sum
    return max_sum / k
nums = [5,5,5,5]
k = 2
print(maxAverage(nums, k))
```
